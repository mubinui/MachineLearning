```python
import numpy as np 
from sklearn.metrics import r2_score 
x = []
y = [100,90,80,60,60,55,40,45,50,55,60,65,70,70,75,76,78,79,90,98,90,100] 
for i in range(1,23):
    x.append(i)
print(len(x), len(y))

mymodel = np.poly1d(np.polyfit(x,y,3)) # returns a function 
myline = np.linspace(1,22,100) 
print(r2_score(y,mymodel(x))) # y and polynomial plot ratio
# greater than 90% so it is a good fit 
```

```python
#Multiple Regression 
import pandas as pd 
df = pd.read_csv('/content/sample_data/cars.csv')
df
```

We dcan predict the CO2 emission of a car based on the size of the engine, but the multiple regression we can throw in more variables, like the weight of the car make the prediction more accurate. 

```python
X = df[['Weight','Volume']] # taking the list of independent values 
Y = df[['CO2']] # Dependent value that we need to predict in the future 
print(X,Y)
```

```python
from sklearn import linear_model
# from the sklearn module we will use LinearRegression() method to create linear regression object 
# the fit() function takes independent and independent values as parameter and fillos the regression  object
# with data that describes the relationship 
```

```python
regr = linear_model.LinearRegression()
regr.fit(X,Y)
```

Lets **predict the CO2 emission** of a car where the weight is 2300kg and the bolume is 1300 cm3

```python
p = regr.predict([[2300,1300]])
print(p)
```

```python
# Lets see all steps in a single block 
# Step 01 
import pandas as pd 
from sklearn import linear_model
# Step 02 
df = pd.read_csv('/content/sample_data/cars.csv')
#Step 03
X = df[['Weight','Volume']]
Y = df[['CO2']]
# Step 04
regr = linear_model.LinearRegression()
regr.fit(X,Y)
#Step05
prediction = regr.predict([[1800,1800]]) # Weight, Volume 
print('Predicted CO2 range',prediction)
#Step06
#Printing the Co-effecient values of the 
print(regr.coef_) # the values tells us if 1 kg weight increases CO2 emmission increases by 0.00755095
# second value tells us if 1cm3 engine size(Volume) increases C02 emmission increases by 0.00780526
```
