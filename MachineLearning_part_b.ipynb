**Data Distributions** Normal Data Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Normal distribution
x = np.random.normal(0.0,10.0,100000)
plt.hist(x,100)
plt.show()
```

```python
#Scatter plot
#Data concentration around a area of a point
x = np.random.uniform(0,10,200)
y = np.random.uniform(80,100,200)
plt.scatter(x,y)
plt.show()
```

```python
import random 
x = []
y = [100,90,80,60,60,55,40,45,50,55,60,65,70,70,75,76,78,79,90,98,90,100] 
for i in range(1,23):
    x.append(i)
print(len(x), len(y))

mymodel = np.poly1d(np.polyfit(x,y,3)) # returns a function 
myline = np.linspace(1,22,100)   

plt.scatter(x,y,color='red')
plt.plot(myline,mymodel(myline))
plt.show()
```
